```python
import json
import glob
import re
```

```python
#reading pavbhaji json file
with open('dataset/pavbhaji.json', 'r') as f:
    data = json.load(f)
```

```python
image_filenames1 = glob.glob('dataset/images/1/*.jpg') # filenames with pavbhaji images
image_filenames0 = glob.glob('dataset/images/0/*.jpg') # filenames with non-pavbhaji images
image_filenames = image_filenames1 + image_filenames0
```

```python
# printing out text content in the json file for analysis
for i, name in enumerate(image_filenames):
    for d in data:
        if d['display_url'].split('/')[-1] == name.split('/')[-1]:
            print (i, d['edge_media_to_caption']['edges'][0]['node']['text'])
            print ('**********************************************************')
            print ()
```

```python
# checking for occurrence of pavbhaji term in all labelled documents
for i, name in enumerate(image_filenames):
    for d in data:
        if d['display_url'].split('/')[-1] == name.split('/')[-1]:
            print (i, re.search('pav.{0,3}bhaji', d['edge_media_to_caption']['edges'][0]['node']['text'].lower()))
```

```python
file_names1_set = {f.split('/')[-1] for f in image_filenames1}
file_names0_set = {f.split('/')[-1] for f in image_filenames0}
```

```python
# writing text content from json file in readable format for inspection

file1 = open('positive_class.txt', 'w')
file2 = open('negative_class.txt', 'w')
for d in data:
    if d['display_url'].split('/')[-1] in file_names1_set:
        file1.write(d['edge_media_to_caption']['edges'][0]['node']['text'] + '\n*********************************************\n\n')
    elif d['display_url'].split('/')[-1] in file_names0_set:
        file2.write(d['edge_media_to_caption']['edges'][0]['node']['text'] + '\n*********************************************\n\n')
file1.close()
file2.close()
```

```python
# printing positions of pavbhaji mentions in every document

for i, name in enumerate(image_filenames):
    for d in data:
        if d['display_url'].split('/')[-1] == name.split('/')[-1]:
            if re.search('pav.{0,3}bhaji', d['edge_media_to_caption']['edges'][0]['node']['text'].lower()):
                print (i, re.search('pav.{0,3}bhaji', d['edge_media_to_caption']['edges'][0]['node']['text'].lower()).span()[0])
            else:
                print (False)
```

```python
# Checking if the pavbhaji class has "pavbhaji" term in the first 150 characters

for i, name in enumerate(image_filenames1):
    for d in data:
        if d['display_url'].split('/')[-1] == name.split('/')[-1]:
            if re.search('pav.{0,3}bhaji', d['edge_media_to_caption']['edges'][0]['node']['text'].lower()):
                print (i, re.search('pav.{0,3}bhaji', d['edge_media_to_caption']['edges'][0]['node']['text'].lower()).span()[0] < 150)
            else:
                print (False)

# the result is hugely positive meaning that in the pavbhaji class, the term "pavbhaji" generally appears early on
```

```python
# checking to see if the term "pavbhaji" appears in the first 40% of the content of the post

for i, name in enumerate(image_filenames1):
    for d in data:
        if d['display_url'].split('/')[-1] == name.split('/')[-1]:
            text = d['edge_media_to_caption']['edges'][0]['node']['text'].lower()
            if re.search('pav.{0,3}bhaji', text):
                print (i, re.search('pav.{0,3}bhaji', text).span()[0]/len(text) < 0.4)
            else:
                print (False)

# the result is hugely positive
```

```python
# checking if the inverse pattern exists in the non-pavbhaji class

for i, name in enumerate(image_filenames0):
    for d in data:
        if d['display_url'].split('/')[-1] == name.split('/')[-1]:
            text = d['edge_media_to_caption']['edges'][0]['node']['text'].lower()
            if re.search('pav.{0,3}bhaji', text):
                print (i, re.search('pav.{0,3}bhaji', text).span()[0]/len(text) < 0.4)
            else:
                print (False)

# the result is mixed, no real pattern
```
